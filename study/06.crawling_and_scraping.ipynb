##### 매개변수를 추가해 요청을 전송하는 방법 p.756

    - 주소에 한글이 들어가있을 경우 딕셔너리를 통해 주소를 encoding 해주어야함

```python
import urllib.request
import urllib.parse
# 데이터 읽어 들이기
API = "http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp" # 데이터 기본 주소
# 매개변수(딕셔너리 자료형)를 URL 인코딩
value = {"stnld" : "108"} # 위 주소(서울경기)를 108로 했으므로 그 다음 강원(109) 등은 바꿔가면서 불러오면 됨
# 여러 데이터를 가져오고 싶을떄는 &를 이용하면 됨 (key=n&key=m 이런식)
# 주소에 한글이 있을 경우 딕셔너리를 통해 바꿔주는과정
params = urllib.parse.urlencode(value) # 매개변수를 URL 인코딩
# params 실행해보면 'stnld=108' 출력
print(params)

# 요청 URL 생성
url = API + "?" + params # URL 리소스 열기
print("url = {}".format(url))

# 다운로드
res = urllib.request.urlopen(url) # .read() 바로 붙여도 됨
data = res.read() # 바이너리 데이터로 읽어 들이기

# 바이너리를 문자열로 변환(HTML 소스 불러오기)
text = data.decode("utf-8") # decode 메소드를 이용한 바이너리를 문자열로 변환
print(text) # 분자열로 출력
```

##### BeautifulSoup로 스크레이핑하기

    - html.parser : 기본 옵션으로 빠르지만 유연하지 못함(단순한 html 문서에서 사용)
    - lxml : 매우 빠르고 유연
    - xml : XML 파일에만 사용
    - html5lib : 매우 느리지만 유연(구조가 복잡한 HTML 문서에 사용)
    - BeautifulSoup() : HtML 구조로 요소를 추출하는 것은 HTML 구조를 하나하나 적는것보다 복잡

```python
from bs4 import BeautifulSoup

url = "https://www.fun-coding.org/crawl_basic2.html" # 데이터를 가져올 주소
res = urllib.request.urlopen(url) # URL 리소스를 열기
data = res.read()
# 바이너리를 문자열로 변환(HTML 소스 불러오기)
text = data.decode("utf-8") # decode 메소드를 이용한 바이너리를 문자열로 변환

# ------------------------------------- 이런식으로 가져온 후 --------------------------------------

# 분석할 HTML
html = text

# HTML 분석하기
soup = BeautifulSoup(html, "html.parser")

# 원하는 부분 추출
h1 = soup.html.body.h1 # <- HTML 태그 구조를 이용하여 접근
p1 = soup.html.body.p
p2 = p1.next_sibling.next_sibling # <- next_sibling 첫 번째 p태그 다음 공백 줄바꿈
                                  #    next_sibling 두 번째 p태그
# P1 = soup.find('p', id = 'ex0') # -> p태그를 찾은 후 id가 ex0인 값을 돌려줌
# 이런식으로 찾아서 출력해도 됨

# 요소의 글자 출력
print("h1 = {}".format(h1)) #h1.string
print("p1 = {}".format(p1)) #p1.string
print("p2 = {}".format(p2)) #p2.string
```

    - previous_sibling : 같은 형제 중 이전 형제를 뽑아냄
    - next_sibling : 같은 형제 중 다음 형제를 뽑아냄
    - 줄이 바뀔 경우 보이지 않지만 줄과 줄 사이에 \n이 들어간 것임
    - 따라서, next_sibling.next_sibling으로 두번 넘겨주어야 원하는 태그 크롤링

$soup . Tag^1 . Tag^2$

##### CSS 선택자로 지정할 수 있는 서식 p.766

기본서식

    - * : 모든 요소 선택
    - <요소 이름> : 요소 이름 기반으로 선택
    - .<클래스 이름> : 클래스 이름 기반으로 선택
    - #<id 이름> : id 속성 기반으로 선택
    
선택자들의 관계를 지정하는 서식 (선택자 : 본인이 표시할 태그 형식)

    - <선택자>, <선택자> : 쉼표로 구분된 여러 개의 선택자를 모두 선택
    - <선택자> <선택자> : 앞 선택자의 후손 중 뒤 선택자에 해당하는 것을 모두 선택
    - <선택자> > <선택자> : 앞 선택자의 자손 중 뒤 선택자에 해당하는 것을 모두 선택
    - <선택자> + <선택자> : 같은 계층에서 바로 뒤에 있는 요소를 선택
    - <선택자> ~ <선택자> : 선택자 1부터 선택자 2까지의 요소를 모두 선택
    
선택자 속성을 기반으로 지정하는 서식 - id나 class외에 속성들

    - <요소>[<속성>] : 해당 속성을 가진 요소를 선택
    - <요소>[<속성> = <값>] : 해당 속성의 값이 지정한 값과 같은 요소를 선택
    - <요소>[<속성> ~= <값>] : 해당 속성의 값이 지정한 값을 단어로 포함하고 있다면 선택
    - <요소>[<속성> |= <값>] : 해당 속성의 값으로 시작하면 선택
    - <요소>[<속성> ^= <값>] : 해당 속성의 값이 지정한 값으로 시작하면 선택
    - <요소>[<속성> $= <값>] : 해당 속성의 값이 지정한 값으로 끝나면 선택
    - <요소>[<속성> *= <값>] : 해당 속성의 값이 지정한 값을 포함하고 있다면 선택
    
위치 또는 상태를 지정하는 서식 - 해당태그(요소)에서만 찾음

    - <요소>:root : 루트 요소
    - <요소>:nth-child(n) : n번째 자식요소 (중간에 p태그가 있다면 그 하위로 넘김)
    - <요소>:nth-last-child : 뒤에서 n번째 자식요소
    - <요소>:nth-of-type(n) : n번째 해당 종류의 요소
    (중간에 태그가 섞여있지 않다면 child랑 of-type이랑 결과값이 같음)

```python
# factorial
def fact(n):
    n += 1
    result = 1
    for i in range(1, n):
        result *= i
    return(result)

fact(6)
```

```python
def fact2(n):
    if n == 1:
        return(1)
    else:
        n *= fact2(n-1)
    return(n)

fact2(6)
```

```python
# 모듈 읽어 들이기------------------------------ 1.
from bs4 import BeautifulSoup
from urllib.request import *
from urllib.parse import *
from os import makedirs
import os.path, time, re

# 이미 처리한 파일인지 확인하기 위한 변수 ------------2.
# A -> B -> C도 있지만 A -> B -> A 일 수도 있으므로 이미 크롤링한 페이지를 담아두기
proc_files = {}

# HTML 내부에 있는 링크를 추출하는 함수--------------3.
def enum_links(html, base):
    soup = BeautifulSoup(html, 'html.parser')
    links = soup.select("link[re!='stylesheet']") # CSS
    links += soup.select("a[href]") # 링크
    result = []
# href 속성을 추출하고, 링크를 절대 경로로 변환--------4.
    for a in links:
        href = a.attrs['href']
        url = urljoin(base, href)
        result.append(url)
    return result

# 파일을 다운받고 저장하는 함수-----------------------5.
def download_file(url):
    o = urlparse(url)
    dir_path = "C:/Users/Affinity/Desktop/study/Module04/modeule04/" # 임의 변경가능
    savepath = dir_path + o.netloc + o.path
    # netloc : 루트주소(www. ~ .com) / path : 루트주소 하위에 있는 폴더 (.com/ ~)
    if re.search(r"/$", savepath): # 폴더라면 index.html
        savepath += "index.html"
    savedir = os.path.dirname(savepath)
    #모두 다운됐는지 확인
    if os.path.exists(savepath): return savepath
    # 다운받을 폴더 생성
    if not os.path.exists(savedir):
        print("mkdir=", savedir)
        makedirs(savedir)
# 파일 다운받기 --------------------------------------6.
    try: # 예외처리 (오류가 발생했을 경우 어떻게 처리할 것인지)
        # 오류가 발생했을 때 뜨는 문구를 지정할 수 있음
        print("download=", url)
        urlretrieve(url, savepath)
        time.sleep(1) # 1초 휴식 --------------------7.
        return savepath
    except:
        print("다운 실패: ", url)
        return None
    
# HTML을 분석하고 다운받는 함수 ------------------------8.
def analyze_html(url, root_url):
    savepath = download_file(url)
    if avepath is None : return
    if savepath in proc_files: return
# 이미 처리됐다면 실행하지 않음--------------------------9.
    proc_files[savepath] = True # 맨앞에 만들었더 딕셔너리
    print("analyze_html=", url)
# 링크 추출------------------------------------------10.
    html = open(savepath, "r", encoding = 'UTF8').read()
    links = enum_links(html, url)
    for link_url in links:
# 링크가 루트 이외의 경로를 나타낸다면 무시---------------11.
        if link_url.find(root_url) != 0:
            if not re.search(r".css$", link_url): continue
        # HTML 이라면
        if re.search(r".(html|htm)$", link_url):
            # 재귀적으로 HTML 파일 분석하기
            analyze_html(link_url, root_url) # 이 부분이 재귀함수 적용 부분!!!!!!!!!!!!!!!
            continue
        # 기타 파일
        download_file(link_url)
```

```python
# naver api 실습
# naver developer -> documents -> 서비스API -> 검색 -> 블로그 -> python
import os
import sys
import datetime
import time
import urllib.request
import json

def get_request_url(url) :
    client_id = "03JhKALKbgFN_vB7x80a"
    client_secret = "UEHri_9L5U"
    print(url)
    rep = urllib.request.Request(url)
    print(rep.get_header)
    rep.add_header("X-Naver-Client-Id", client_id)
    rep.add_header("X-Naver-Client-Secret", client_secret)
    try:
        response = urllib.request.urlopen(rep)
        if response.getcode() == 200:
            print("[%s] Url request success" % datetime.datetime.now())
            return response.read().decode("utf-8")
        # return 값은 정상 처리되엇을때 결과를 디코드(한글로)해서 보여주도록
    except Exception as e: # 발생한 Error를 e에 저장하여 확인할 수 있음
        print(e)
        print("[%s] Error for URL : %s" % (datetime.datetime.now(), url))
        return None
    
def getNaverSearchResult(sNode, search_text, page_start, display):
    # 6.예시의 호출부분에 주소를 분할에서 지정해주는 과정
    base = "https://openapi.naver.com/v1/search"
    node = "/%s.json" % sNode # 블로그인지 뉴스인지 카페인지 등등
    params = "?query=%s&start=%s&display=%s" % (urllib.parse.quote(search_text),
                                               page_start, display)
    # quote 함수는 받은 문자를 인코딩한 문자로 돌려준다
    url = base + node + params
    
    getData = get_request_url(url)
    
    if (getData == None):
        return None
    else:
        return json.loads(getData)

# 비로그인 방식으로 뉴스나 블로그 검색한 결과를 돌려받는 방식
def getPostData(post, jsonResult):
    title = post["title"]
    description = post["description"]
    link = post["link"]
    org_link = post["originallink"]
    
    pDate = datetime.datetime.strptime(post["pubDate"],
                                       "%a, %d %b %Y %H:%M:%S +0900")
    # strptime : 날짜, 시간형식의 문자열을 datetime으로 만들 때
    pDate = pDate.strftime('%Y-%m-%d %H:%M:%S')
    # strftime : 시간을 문자열로 출력할 때
    jsonResult.append({"title" : title, "description" : description,
                      "org_link" : org_link, "link" : link, "pDate" : pDate})
    return

def main() :
    jsonResult = []
    sNode = "news" # blog, cafe, article 등 검색할거 작성
    search_text = "빅데이터"
    display_count = 100
    page_start = 1
    
    jsonSearch = getNaverSearchResult(sNode, search_text, page_start, display_count)
    
    while ((jsonSearch != None) and (jsonSearch['display'] != 0)):
        for post in jsonSearch["items"] :
            getPostData(post, jsonResult)
        nStart = jsonSearch["start"] + jsonSearch["display"]
        # 1부터 100개씩 보였으면 다음번호는 101번이 되게
        jsonSearch = getNaverSearchResult(sNode, search_text, nStart, display_count)
        
    # 출력한 값을 파일로 저장하기
    with open('%s_naver_%s.json' % (search_text, sNode), 'w',
             encoding = "utf-8") as outfile :
        retJson = json.dumps(jsonResult, indent = 4, sort_keys = True,
                            ensure_ascii = False)
        # indent : 들여쓰기 공백 4개(\t도 가능)
        # ensure_ascii = False 인 이유 : 우리는 한글이 들어가있기 때문
        outfile.write(retJson)
    print("%s_naver_%s.json SAVED" % (search_text, sNode))
    
if __name__ == "__main__" :
    main()
```
