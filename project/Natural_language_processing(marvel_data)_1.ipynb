```python

def get_data(test_url, total_count):
    review_list = []
    for i in range(1, int(total_count / 10) + 1):
#         headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36'}
        url = test_url + '&page=' + str(i)
#         html = requests.get(url, headers = headers).text
        resp = requests.get(url)
        html = BeautifulSoup(resp.content, 'html.parser')
        score_result = html.find('div', {'class': 'score_result'})
        lis = score_result.findAll('li')
        for li in lis:        
            review_text = li.find('p').getText()
            review_list.append(review_text)
        print(i)
        # 간단하게 프린트만 했습니다.
    return review_list
```

```python
from konlpy.tag import Hannanum
han = Hannanum()

movie_li = [captain, endgame, infinitywar, spider]
result = []

for movie in movie_li:
    for line in movie:
        mallist = han.pos(line)
        r=[]
        for word in mallist:
            if not word[1] in ["J","E","S"]: 
                r.append(word[0])
        rl = (" ".join(r)).strip()
        result.append(rl)
```

```python
with open("review.nip","w",encoding="utf-8") as fp:
    fp.write("\n".join(result))
```

```python
from gensim.models import word2vec

wData = word2vec.LineSentence("review.nip")
wModel = word2vec.Word2Vec(wData, size=10000, window=5, min_count=2, sg=1)
wModel.save("review.model")
print("Word2Vec Modeling finished")
```
