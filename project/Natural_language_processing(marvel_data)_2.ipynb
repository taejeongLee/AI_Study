```python
def get_data(movie_url, total_count):

    text_list = list()
    
    for i in range(1, int(total_count / 10) + 1):
            url = movie_url + '&page=' + str(i)
            resp = requests.get(url)
            html = BeautifulSoup(resp.content, 'html.parser')
            score_result = html.find('div', {'class': 'score_result'})
            lis = score_result.findAll('li')
            
            for li in lis:  
                review_text = li.find('p').getText()
                if(review_text[:3] == "관람객"):
                    review_text = review_text[3:]
                text_list.append(review_text)
            print("페이지 {}/{}".format(i,int(total_count / 10) + 1))
    return text_list
```

```python
test_url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=136315&type=after"
infinitywar_list = get_data(test_url,total_count)
```

```python
import pickle
with open("infinityWar_review.txt","wb") as f:
    pickle.dump(infinitywar_list, f)
f.close()
```

## 형태소 분석

```python
from konlpy.tag import Hannanum
han = Hannanum()

result = []

for line in infinitywar_list:
    mallist = han.pos(line)
    r=[]
    for word in mallist:
        if not word[1] in ["J","E","S"]: 
            r.append(word[0])
    rl = (" ".join(r)).strip()
    result.append(rl)
```

```python
result
```

```python
with open("infinitywar.nip","w",encoding="utf-8") as fp:
    fp.write("\n".join(result))
```

```python
from gensim.models import word2vec

wData = word2vec.LineSentence("infinitywar.nip")
wModel = word2vec.Word2Vec(wData, size=5000, window=10, min_count=2, sg=1)
wModel.save("infinitywar.model")
print("Word2Vec Modeling finished")
```
